```python
import ast
def convert1(obj):
  l = []
  for i in ast.literal_eval(obj):   #[{}{}{}]
    l.append(i['name'])
  return l

def convert2(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter == 3:
      break
    l.append(i['name'])
    counter = counter + 1
  return l

def convert3(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i['job'] == "Director":
      l.append(i['name'])
  return l
```

```python
df['genres']= df['genres'].apply(convert1)
df['keywords']= df['keywords'].apply(convert1)
df['cast'] = df['cast'].apply(convert2)
df['crew'] = df['crew'].apply(convert3)
df.head()
```

```python
# df['overview'] = df['overview'].apply(lambda x : x.split())
# Handle potential float values by converting them to strings first
df['overview'] = df['overview'].astype(str).apply(lambda x: x.split())
```

```python
df['genres']= df['genres'].apply(lambda x : [i.replace(" ","") for i in x])
df['keywords']= df['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x : [i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
df['tags']= df['overview']  + df['keywords'] + df['cast'] + df['crew'] + df['genres']
df.head()
```

```python
new_df = df[['movie_id','title','tags']]
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
list(similarity[0])
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key=lambda x : x[1] )[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df["title"] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True, key=lambda x : x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```
